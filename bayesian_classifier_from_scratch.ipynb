```python
def MagrginProp(p_labels,p_feature,keys_feature):
    # let's compute marginal probabilities of labels values
  p_feature = np.sum(p_labels * p_feature, axis=1)
  # we create a pandas dataframe to visualize the table more familiar
  df_p_feature = pd.DataFrame(p_feature, keys_feature)

  # possibilities of buying values
  return df_p_feature
```

```python
def JointProp(train_dataset,featurecolumn,labelcolumn="Car Acceptability"):
    # unique values of feature column
  keys_feature = np.array(train_dataset[featurecolumn].unique())

  # unique values of label column
  keys_labels = np.array(train_dataset[labelcolumn].unique())

  # let's initiate a matrix to hold counts 
  p_feature_given_label = np.zeros((len(keys_feature), len(keys_labels)))

  # zip only feature and clazz label
  for i in zip(train_dataset[featurecolumn], train_dataset[labelcolumn]):
      # get index of current feature value
      index_feature = np.where(keys_feature == i[0])
      
      # get index of current label value
      index_label = np.where(keys_labels == i[1])
      
      # increment 1 the count of the intersection of feature x label pair.
      p_feature_given_label[[index_feature], [index_label]] += 1

  # we create a pandas dataframe to visualize the table more familiar
  df_p_feature_given_label = pd.DataFrame(p_feature_given_label, keys_feature, keys_labels)
  # let's normalize the possibilities
  p_feature_given_label = p_feature_given_label / np.sum(p_feature_given_label, axis=0, keepdims=True)

  # we create a pandas dataframe to visualize the table more familiar
  df_p_feature_given_label = pd.DataFrame(p_feature_given_label, keys_feature, keys_labels)

  # possibilities of buying values given class
  return df_p_feature_given_label 
```
